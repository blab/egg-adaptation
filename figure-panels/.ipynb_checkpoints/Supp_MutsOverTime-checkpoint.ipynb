```python
import json
import logomaker
from collections import Counter
from augur.utils import json_to_tree
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.lines import Line2D
from matplotlib.gridspec import GridSpec
import pandas as pd
import numpy as np
```

```python
adaptive_muts_by_virus = {'h3n2':{138:['S'], 156: ['Q','R'], 160: ['K', 'I'], 186: ['V', 'N'], 
                                  190: ['G', 'V','N'], 193: ['R'], 194: ['I', 'P'], 195: ['Y'], 
                                  203: ['I'], 219: ['Y/F'], 225: ['G', 'N'], 226: ['I'], 
                                  246: ['K', 'S', 'H', 'T']}, 
                          'h1n1pdm': {127: ['E'], 187: ['V', 'N', 'T'], 191: ['I'], 
                                      222: ['G', 'N'], 223: ['R'], 224: ['K']}, 
                          'vic': {141: ['R'], 196: ['S', 'D', 'K', 'T'], 198: ['I', 'A', 'N']}, 
                          'yam': {141: ['R'], 195: ['S', 'D', 'K'], 197: ['I', 'A', 'N', 'P']}}
```

```python
def get_viruses_w_adaptive_muts(virus):
    """
    Find the number of egg-passaged viruses with at least one adaptive mut
    """
    
    curated_mut_file = f"../egg-mutation-analysis/egg-muts-by-strain/{virus}_ha_curated-egg-muts.json"
    
    with open(curated_mut_file) as json_handle:
        egg_mut_info = json.load(json_handle)
    
    # count total number of egg passaged viruses
    total_egg_strains = 0
    # and number with at least one adaptive mut
    strains_w_mut = 0

    for strain, muts in egg_mut_info.items():
        total_egg_strains+=1
        has_mut=False
        for g, m_list in muts.items():
            if g=='HA1':
                for m in m_list:
                    if has_mut == False:
                        m_pos = int(m[:-1])
                        if m_pos in adaptive_muts_by_virus[virus].keys():
                            if m[-1] in adaptive_muts_by_virus[virus][m_pos]:
                                strains_w_mut+=1
                                has_mut=True

    
                    
    proportion_w_mut = strains_w_mut/total_egg_strains
    
    return proportion_w_mut
        
```

```python
virus_cmap = {'B/Yam': '#fbba0a', 'B/Vic': '#FF4500', 'A/H3N2': '#6C3BAA', 'A/H1N1pdm': '#45b6fe'}
```

```python
def plot_proportion_w_adaptive_mut(filename=None):
    """
    """
    #plot styling
    sns.set_style("white")
    plt.rcParams['figure.dpi']= 300
    plt.rcParams['font.family'] = 'Arial'
    
    viruses = ['h3n2', 'h1n1pdm', 'vic', 'yam']
    prettify_names = {'h3n2': 'A/H3N2', 'h1n1pdm': 'A/H1N1pdm', 'vic': 'B/Vic', 'yam': 'B/Yam'}
    
    data_to_plot = {}
    for v in viruses:
        data_to_plot[prettify_names[v]] = get_viruses_w_adaptive_muts(v)

    
    fig, ax = plt.subplots(figsize=(5,3))
    
    for virus, p in data_to_plot.items():
        ax.bar(x=virus, height=p, color=virus_cmap[virus])
    
    
    ax.set_ylabel('Proportion with at least one\nadaptive mutation in HA1')
    ax.set_xlabel('')
    
    sns.despine()
```

```python
plot_proportion_w_adaptive_mut()
```
