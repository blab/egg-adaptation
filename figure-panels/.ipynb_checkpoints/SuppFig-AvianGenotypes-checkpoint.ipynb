```python
import json
from collections import Counter
from Bio import SeqIO
from Bio.SeqFeature import FeatureLocation
from augur.utils import json_to_tree
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import pandas as pd
```

## Supplementary Fig: show genotype of avian H1 and H3 at egg-adaptive sites, from builds using human ref for alignment

```python
# egg-adaptive aas (residue is mut in 3% of more of strains and AA occurs at least 5 times)
egg_adaptive_aas = {'h3n2': {138: ['S'], 156: ['Q', 'R'], 160: ['K', 'I'], 
                             186: ['V', 'N', 'R', 'D', 'S', 'A', 'Y'], 
                             190: ['G', 'V', 'N', 'E', 'I'], 
                             194: ['P', 'I', 'M', 'V'], 195: ['Y'], 203: ['I'], 
                             219: ['Y', 'F'], 225: ['G', 'N'], 
                             226: ['I'], 246: ['K','S', 'H', 'T']}, 
                    'h1n1pdm': {127:['S'], 187: ['V', 'N', 'T'], 191: ['I'], 
                                222: ['G', 'N'], 223: ['R']}
                  }
```

```python
# egg-adaptive sites to look for
egg_sites = {'h3n2': list(egg_adaptive_aas['h3n2'].keys()), 
             'h1n1pdm': list(egg_adaptive_aas['h1n1pdm'].keys())
                  }
```

```python
def get_sample_date(virus):
    """
    Because date is nto supplied in alignment file, get date for each sequence from metadata file
    """
    ha_type = virus[0:2].upper()
    metadata_file = f'../nextstrain_builds/avian/results/{ha_type}/metadata.tsv'
    
    metadata = pd.read_csv(metadata_file, sep='\t')
    
    # map strain name to year
    date_mapper = {r['strain']: r['date'] for i,r in metadata.iterrows()}
    year_mapper = {k: str(v)[:4] for k,v in date_mapper.items()}
    
    return year_mapper
```

```python
def get_avian_genotypes(virus):
    """
    Read in the alignment and determine genotype at each of the positions
    """
    ha_type = virus[0:2].upper()
    # get year for each sample
    # maps strain name to year
    year_mapper = get_sample_date(virus)
    
    # get location of HA1
    tree_path= f'../nextstrain_builds/avian/auspice/{ha_type}.json'
    
    #read in the tree file to get the genome annotations
    with open(tree_path, 'r') as f:
        tree_json = json.load(f)
    
    ha1_location = tree_json['meta']['genome_annotations']['HA1']
    ha1_feature_loc= FeatureLocation(ha1_location['start']-1, ha1_location['end'])
    
    alignment_file = f'../nextstrain_builds/avian/results/{ha_type}/aligned.fasta'
    
    # record the genotype at each egg site
    genotypes = {x:[] for x in egg_sites[virus]}
    # keep track of genotype by year
    genotype_by_year = {x:{str(y):[] for y in range(1960,2025)} for x in egg_sites[virus]}
    
    for record in SeqIO.parse(alignment_file, "fasta"):
        # get year
        y = year_mapper[record.id]
        # get ha1` seq
        ha1_seq = ha1_feature_loc.extract(record.seq)
        # replace all - with N, so I can translate
        # will result in translating Xs but I can ignore these
        ha1_seq = ha1_seq.replace('-', 'N')
        ha1_aa_seq = ha1_seq.translate()
        # sites are 1-based but indexing is 0
        for x in egg_sites[virus]:
            genotypes[x].append(ha1_aa_seq[x-1])
            genotype_by_year[x][y].append(ha1_aa_seq[x-1])

    
    genotype_counts = {x:Counter(g) for x,g in genotypes.items()}
     
    genotype_by_year_counts = {}
    for s, yg in genotype_by_year.items():
        genotype_by_year_counts[s] = {x:Counter(g) for x,g in yg.items()}
    
    return genotype_by_year_counts
```

```python
def predominant_avian_genotypes(virus, site):
    """
    Return any genotype that reaches 33% or more at the site
    """
    
    genotype_by_year_counts = get_avian_genotypes(virus)
    
    genotype_at_site_by_year_counts = genotype_by_year_counts[site]
    
    predominant_genotypes_at_site = []
    
    for y, d in genotype_at_site_by_year_counts.items():
        year_total = sum(d.values())
        # must have at least 3 seqs in year
        if year_total>=5:
            for g, c in d.items():
                if c >= 0.33*year_total:
                    if g not in predominant_genotypes_at_site:
                        predominant_genotypes_at_site.append(g)
                    
                    
    return predominant_genotypes_at_site
```

```python
def plot_genotype_over_time(virus, genotype_info, ax, include_y_ticks=True):
    """
    Plot the genotype over time at a given position
    """
    plt.rcParams['font.family'] = 'Arial'
    # map each aa to a different color
    all_amino_acids = list("ACDEFGHIKLMNPQRSTVWY")

    color_palette = sns.color_palette("tab20", 20)

    cmap = {aa: color for aa, color in zip(all_amino_acids, color_palette)}
    
    aa_by_year = []
    
    for y, gc in genotype_info.items():
        total_in_year = sum(gc.values())
        # require at least 5 seqs in the year
        if total_in_year>=5:
            for g,c in gc.items():
                # don't include Xs
                if g!='X':
                    aa_by_year.append({'year':int(y), 'aa': g, 'y': c/total_in_year})
            
    df = pd.DataFrame(aa_by_year)
    
    sns.lineplot(data=df, x='year', y='y', hue='aa', palette=cmap, ax=ax, legend=False)
    
    ax.set_ylim(0, 1.02)
    
    ax.set_xlabel('')
    ax.set_ylabel('')
    
    if include_y_ticks==False:
        ax.set_yticklabels([])
    
    sns.despine()
    
    return ax
```

```python
def plot_aa_legend(ax):
    """
    Make the legend showing what color each aa is 
    """
    plt.rcParams['font.family'] = 'Arial'
    # map each aa to a different color
    all_amino_acids = list("ACDEFGHIKLMNPQRSTVWY")

    color_palette = sns.color_palette("tab20", 20)

    cmap = {aa: color for aa, color in zip(all_amino_acids, color_palette)}
    
    # make a legend line for each aa
    aa_color_lines = [Line2D([0], [0], color=cmap[aa], lw=2) for aa in all_amino_acids]
    
    ax.legend(aa_color_lines, all_amino_acids, ncol=2, labelspacing=0.25, title='Amino acid', fontsize=8, 
              bbox_to_anchor=(1, 0.8))
            
    ax.yaxis.set_visible(False) 
    ax.spines['left'].set_color('white')
    
    ax.xaxis.set_visible(False) 
    ax.spines['bottom'].set_color('white')
    
    sns.despine()
            
    return ax
```

```python
# Function to load and resize an image
def load_image(image_path, zoom=0.1):
    image = plt.imread(image_path)
    return OffsetImage(image, zoom=zoom)

def write_genotypes(virus, site, ax):
    """
    Write the egg-adaptive genotypes and then the observed avian genotypes at these positions
    """
    
    egg_image = load_image('Supp_avian/egg2.jpeg', zoom=0.06)
    duck_image = load_image('Supp_avian/duck.jpeg', zoom=0.04)
    
    ax.axis('off')  
    # label the site
    ax.text(0.05, 0.7, site, fontsize=18, va='center', transform=ax.transAxes)

    # Add the text for egg-adaptive muts
    egg_image_box = AnnotationBbox(egg_image, (0.02, 0.4), frameon=False, xycoords='axes fraction')
    ax.add_artist(egg_image_box)
    egg_genotypes = ', '.join(egg_adaptive_aas[virus][site])
    ax.text(0.15, 0.4, egg_genotypes, fontsize=12, va='center', transform=ax.transAxes)

    # Add the duck genotypes
    duck_image_box = AnnotationBbox(duck_image, (0.02, 0.1), frameon=False, xycoords='axes fraction')
    ax.add_artist(duck_image_box)
    avian_genotypes = predominant_avian_genotypes(virus, site)
    avian_geno_text = ', '.join(avian_genotypes)
    ax.text(0.15, 0.1, avian_geno_text, fontsize=12, va='center', transform=ax.transAxes)
    
    return ax
```

```python
def plot_avian_genotypes_supp(filename=None):
    """
    Plot avian genotypes at each egg-adaptive site for H1N1pdm and H3N2
    """
    
    #plot styling
    sns.set_style("white")
    plt.rcParams['figure.dpi']= 300
    plt.rcParams['font.family'] = 'Arial'
    
    fig = plt.figure(figsize=(12, 10))
    # plot H1 sites (5) on first row, then the aa legned
    # plot H3 sites (12) on the second two rows
    # blank row in between for spacing
    gs = GridSpec(8, 6, width_ratios=[1,1,1,1,1,1], height_ratios=[1,1,0.4, 1,1,0.1, 1,1])
    
    # get H1 avian genotypes 
    H1_genotype_by_year_counts = get_avian_genotypes('h1n1pdm')
    
    for i, s in enumerate(egg_sites['h1n1pdm']):
        ax_h1_text = fig.add_subplot(gs[0, i])
        write_genotypes('h1n1pdm', s, ax_h1_text)
        
        ax_h1 = fig.add_subplot(gs[1, i])
        # include y-tick labels
        if i==0:
            plot_genotype_over_time('h1n1pdm', H1_genotype_by_year_counts[s], ax_h1)
            # label H1
            ax_h1_text.text(-0.3, 1.05, 'Avian H1', fontsize=20, transform=ax_h1_text.transAxes)

        # don't include y-tick labels
        else:
            plot_genotype_over_time('h1n1pdm', H1_genotype_by_year_counts[s], ax_h1, False)

            
    # plot AA legend
    ax_leg = fig.add_subplot(gs[0:1, 5])
    plot_aa_legend(ax_leg)
        


    # get H3 avian genotypes 
    H3_genotype_by_year_counts = get_avian_genotypes('h3n2')
    
    for i, s in enumerate(egg_sites['h3n2']):
        if i<=5:
            ax_h3_text = fig.add_subplot(gs[3, i])
            write_genotypes('h3n2', s, ax_h3_text)
            ax_h3 = fig.add_subplot(gs[4, i])
            # include y-tick labels on left most plot only
            if i==0:
                plot_genotype_over_time('h3n2', H3_genotype_by_year_counts[s], ax_h3)
                # label H3
                ax_h3_text.text(-0.3, 1.05, 'Avian H3', fontsize=20, transform=ax_h3_text.transAxes)
            else:
                plot_genotype_over_time('h3n2', H3_genotype_by_year_counts[s], ax_h3, False)
            # label residue site

            
        else:
            ax_h3_text = fig.add_subplot(gs[6, 6-i])
            write_genotypes('h3n2', s, ax_h3_text)
            ax_h3 = fig.add_subplot(gs[7, 6-i])
            # include y-tick labels on left most plot only
            if i==6:
                plot_genotype_over_time('h3n2', H3_genotype_by_year_counts[s], ax_h3)
            else:
                plot_genotype_over_time('h3n2', H3_genotype_by_year_counts[s], ax_h3, False)

    
    if filename:
        fig.savefig(filename, dpi=300, bbox_inches='tight', transparent=False, facecolor='white')

    
```

```python
plot_avian_genotypes_supp(filename='Supp_avian/Supplement_aviangenotypes.png')
```
