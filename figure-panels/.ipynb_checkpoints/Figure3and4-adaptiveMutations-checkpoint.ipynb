# Figure 3 and 4: Egg-adaptive mutations occur near the receptor-binding pocket of HA1 (fig3), and in NP (fig4)

```python
def get_length_gene(virus, segment):
    """
    Get the length of each gene or subunit on the segment
    """
    
    
    reference_file = f'../nextstrain_builds/egg-enriched/config/{virus}/{segment}/genemap.gff'
    
    gene_aa_pos = {}
    
    with open(reference_file, 'r') as gff_handle:
        gff_lines = gff_handle.readlines()
        for line in gff_lines:
            if 'gene' in line.split('\t'):
                gene_name = line.split('\t')[8].split('gene_name=')[1].split('"')[1]
                gene_pos = [int(line.split('\t')[3])-1, int(line.split('\t')[4])]
                gene_aa_len = int((gene_pos[1]-gene_pos[0])/3)
                gene_aa_pos[gene_name] = gene_aa_len
    
    return gene_aa_pos
    
```

```python
def get_muts_over_threshold(virus, segment, threshold):
    """
    For the given segment, return amino acid positions with greater than the 
    threshold proportion of strains mutated during egg-passaging
    """
    
    curated_mut_file = f"../egg-mutation-analysis/egg-muts-by-strain/{virus}_{segment}_curated-egg-muts.json"
    
    with open(curated_mut_file) as json_handle:
        egg_mut_info = json.load(json_handle)
    
    
    # initialize a count of 0 at each position
    gene_aa_lens = get_length_gene(virus, segment)
    mut_pos_counts = {g:{x:0 for x in range(1,v+1)} for g,v in gene_aa_lens.items()}
    
    # count total number of egg-passaged strains for this segment
    total_strains = 0

    for strain, muts in egg_mut_info.items():
        total_strains+=1
        for g, m_list in muts.items():
            if g!='nuc':
                for m in m_list:
                    m_pos = int(m[:-1])
                    mut_pos_counts[g][m_pos]+=1
    
    # turn counts into proporitons
    mut_pos_proportions = {}
    # return a list of all sites that are mutated in over X=threshold number of strains
    over_threshold = {g:[] for g in gene_aa_lens.keys()}
    for g, pc in mut_pos_counts.items():
        mut_pos_proportions[g] = {}
        for p, c in pc.items():
            prop = c/total_strains
            mut_pos_proportions[g][p] = prop
            if prop >=threshold:
                over_threshold[g].append(p)
                
    
    return over_threshold, mut_pos_proportions, mut_pos_counts, total_strains
        
```

```python
# position of RBD (relative to HA1 numbering
rbd = {'h3n2': np.arange(120,269+1), 
       'h1n1pdm': np.arange(109,259+1), 
       'vic': np.arange(116,276+1), 
       'yam': np.arange(116,275+1)}
```

```python
np_residue_cmap = {'h3n2': {290: '#FD3855', 384: '#00a18a'}, 
                   'h1n1pdm': {101: '#00b4d8', 102: '#90e0ef', 375: '#00a18a'}}
```

```python
# save a file with the sites with egg muts exceding 3%, and what color to use for each site
save_adaptive_sites = {}

viruses = ['h3n2', 'h1n1pdm', 'vic', 'yam']
segments = ['pb1', 'pb2', 'pa', 'ha', 'np', 'na', 'mp', 'ns']
for v in viruses:
    adaptive_sites_this_virus = {}
    for s in segments:
        over_threshold, mut_pos_proportions, mut_pos_counts, total_strains = get_muts_over_threshold(v, s, 0.03)
        for g, ms in over_threshold.items():
            if len(ms)!=0:
                if s=='ha':
                    adaptive_sites_this_virus[g] = {x:ha_residue_cmap[v][x] for x in ms}
                elif s=='np':
                    adaptive_sites_this_virus[g] = {x:np_residue_cmap[v][x] for x in ms}
    save_adaptive_sites[v] = adaptive_sites_this_virus
            
json_object_to_save = json.dumps(save_adaptive_sites, indent=2)

# Write the adaptive mutations at this threshold and 
with open(f"Figure3/egg-adaptive-muts-threshold_3percent.json", "w") as outfile:
    outfile.write(json_object_to_save)
```

```python
def plot_egg_adaptive_muts(virus, segment, threshold, ax, colorby='all', filename=None):
    """
    Plot number of muts that occur at each residue during egg-passaging, 
    for each segment that has any sites exceeding 0.03 (HA for all viruses, NP also for H3N2 and H1N1pdm).
    Color and label residues that exceed 0.03
    """
    
    # get mutation count and proportion info for this virus, segment
    over_threshold, mut_pos_proportions, mut_pos_counts, total_strains = get_muts_over_threshold(virus, segment, threshold)
        
    #plot styling
    sns.set_style("white")
    plt.rcParams['font.family'] = 'Arial'
    plt.rcParams['axes.labelsize'] = 12
    plt.rcParams['axes.labelweight'] = 'normal'
    plt.rcParams['axes.titlesize'] = 12
    plt.rcParams['xtick.labelsize'] = 12
    plt.rcParams['ytick.labelsize'] = 12
    plt.rcParams['figure.dpi']= 300

    # resi numbers will be X axis and propportion of egg strains with mut will be Y axis
    residue_numbers = []
    egg_mut_proportions = []

    
    # if the segment is HA, there will be separate numbering for SigPep, HA1 and HA2
    # label both the HA1 numbering and the HA numbering
    if segment=='ha':
        for sp_x, c in mut_pos_proportions['SigPep'].items():
            residue_numbers.append(sp_x)
            egg_mut_proportions.append(c)
            
        sp_end = max(mut_pos_proportions['SigPep'].keys())
        for ha1_x, c in mut_pos_proportions['HA1'].items():
            residue_numbers.append(ha1_x+sp_end)
            egg_mut_proportions.append(c)
            
        ha1_end = max(mut_pos_proportions['HA1'].keys())
        for ha2_x, c in mut_pos_proportions['HA2'].items():
            residue_numbers.append(ha2_x+ha1_end)
            egg_mut_proportions.append(c)
            
    else:
        for x, c in mut_pos_proportions[segment.upper()].items():
            residue_numbers.append(x)
            egg_mut_proportions.append(c)
            
    # color points above threshold frequency 
    color_list = []
      
    for i, x in enumerate(egg_mut_proportions):
        if x >= threshold:
            if segment=='ha':
                ha1_numbering = i-sp_end+1
                color_list.append(ha_residue_cmap[virus][ha1_numbering])
            elif segment=='np':
                color_list.append(np_residue_cmap[virus][i+1])

            else:
                color_list.append('red')
        else:
            color_list.append('#d3d3d3')

            
    # make dual y-axes with freq and count
    def freq2count(x):
        return x*total_strains
    def count2freq(x):
        return x/total_strains

            
    # make lollipop plot at each codon position, where stem is black and point is colored if there are 
    # egg-adaptive muts at that position
    ax.bar(x=residue_numbers, height=egg_mut_proportions, color='black', edgecolor='black',linewidth=1, zorder=2)
    ax.scatter(x=residue_numbers, y=egg_mut_proportions, color=color_list, s=50, 
               edgecolor='black', linewidth=0.75, zorder=3)
    # draw dashed line at threshold
    ax.axhline(threshold, 0,1, linestyle='--', color='#ff6464')
    
    secax = ax.secondary_yaxis('right', functions=(freq2count, count2freq))
#     secax.set_ylabel('Number of egg-passaged strains')
    secax.set_ylabel('Count')
        
#     ax.set_ylabel('Proportion of egg-passaged strains\nwith mutation at residue', size=14)
    ax.set_ylabel('Proportion', size=14)
        
    # make numbering start at 1
    ax.xaxis.set_ticks(np.arange(1, max(residue_numbers), 100))
    # label x-axis
    ax.set_xlabel(f'{segment.upper()} amino acid position')
        
    # for HA, want to label HA1 and RBD
    if segment=='ha':
        # make secondary X-axis for HA1 numbering
#         secax_x = ax.twiny()  
        secax_x = ax.secondary_xaxis('bottom')
        secax_x.spines['bottom'].set_position(('outward', 40))
        # Align it with the main X-axis
        secax_x.set_xlim(ax.get_xlim())  
        # where HA1 starts, relative to HA-numbering
        ha1_start = sp_end+1
        # tick start and end
#         secax_x.set_xticks([i for i in range(ha1_start, ha1_end + 1, 100)])
        secax_x.set_xticks([ha1_start, ha1_end])
        # label according to HA1 coordinates
        secax_x.set_xticklabels([1, ha1_end-ha1_start])
        # limit the secondary axis to only the length of HA1
        secax_x.spines['bottom'].set_bounds(ha1_start, ha1_end)
        secax_x.spines['bottom'].set_color('#9d9d9d')
        secax_x.xaxis.set_tick_params(direction='in')
        secax_x.spines['bottom'].set_linewidth(2)
        secax_x.xaxis.set_tick_params(width=2)
        secax_x.set_xlabel('HA1')

        secax_x.xaxis.set_label_coords(0.33, -0.23, transform=ax.transAxes)
        secax_x.xaxis.label.set_color('#9d9d9d')
        secax_x.tick_params(axis='x', colors='#9d9d9d')
        
        # mark where RBD is
        rbd_pos = [x+ha1_start for x in rbd[virus]]
        ax.fill_between(rbd_pos, 0, ax.get_ylim()[1],
                        color=cmap[virus]['shading'], alpha=0.1, linewidth=0, zorder=0)
        ax.text(rbd_pos[-1]+2, 0.8, 'RBD', color=cmap[virus]['rbd_text'], fontsize=12, transform=ax.get_xaxis_transform())
        
        # legend on top plot only
        if virus== 'h3n2':
            legend_elements = [Line2D([0], [0], marker='o', color='w', markerfacecolor='#fb8500', markersize=8, lw=0, label='130-loop'),
                               Line2D([0], [0], marker='o', color='w', markerfacecolor='#00a18a', markersize=8, lw=0, label='150-loop'),
                               Line2D([0], [0], marker='o', color='w', markerfacecolor='#9C3587', markersize=8, lw=0, label='190-helix'),
                              Line2D([0], [0], marker='o',  color='w',markerfacecolor='#00b4d8', markersize=8, lw=0, label='220-loop'),
                              Line2D([0], [0], marker='o',  color='w',markerfacecolor='#FD6484', markersize=8, lw=0, label='other')]
            
            
            
            ax.legend(handles=legend_elements, bbox_to_anchor=(1,1), bbox_transform=ax.transAxes)



    
    prettify_name = {'h3n2': 'A/H3N2', 'h1n1pdm': 'A/H1N1pdm', 'vic':'B/Vic', 'yam':'B/Yam'}
    if segment=='np' and virus=='h1n1pdm':
        ax.text(0.02, 0.9, f'{prettify_name[virus]}\n{segment.upper()}', fontsize=16, transform=ax.transAxes)
    else:
        ax.text(0.02, 0.8, f'{prettify_name[virus]}\n{segment.upper()}', fontsize=16, transform=ax.transAxes)
        
    sns.despine()

    

    
    
#     if filename:
#         fig.savefig(filename, dpi=300, bbox_inches='tight', transparent=False, facecolor='white')
        
    return ax
```

```python
def make_logo_plot(virus, segment, threshold, ax):
    """
    Make a logo plot of the amino acids seen in egg-passaging mutations
    """
    
    # get mutation count and proportion info for this virus, segment
    over_threshold, mut_pos_proportions, mut_pos_counts, total_strains = get_muts_over_threshold(virus, segment, threshold)
    
    # adaptive muts are in HA1 or NP
    if segment=='ha':
        gene = 'HA1'
    elif segment=='np':
        gene = 'NP'
    
    curated_mut_file = f"../egg-mutation-analysis/egg-muts-by-strain/{virus}_{segment}_curated-egg-muts.json"
    
    with open(curated_mut_file) as json_handle:
        egg_mut_info = json.load(json_handle)
    
    
    # initialize dict to keep track of all AA muts seen at each position
    aa_muts = {x:[] for x in over_threshold[gene]}
    
    # list of all adaptive residues
    resis = list(over_threshold[gene])
    

    for strain, muts in egg_mut_info.items():
        for g, m_list in muts.items():
            if g==gene:
                # see if mut is at one of the "adaptive" sites, and keep track of what the AA mut is
                for m in m_list:
                    m_pos = int(m[:-1])
                    if m_pos in resis:
                        aa_mut = m[-1]
                        aa_muts[m_pos].append(aa_mut)
                        
    # make df with proportion of each AA at position
    all_possible_aas = [x for x in 'ACDEFGHIKLMNPQRSTVWY']
    to_df = []
    for pos, aas in aa_muts.items():
        total_muts_at_pos = len(aas)
        aa_counts = Counter(aas)
        # get count of each possible aa (to make matrix for logo plotter)
        count_all_possible_aas = [aa_counts.get(a,0) for a in all_possible_aas]
        # normalize by number of muts at this position
        prop_all_possible_aas = [x/total_muts_at_pos for x in count_all_possible_aas]
        to_df.append({'pos': pos, **dict(zip(all_possible_aas,prop_all_possible_aas))})
    
    df = pd.DataFrame(to_df).set_index('pos')
    
    # Fill to target length 
    if segment=='ha':
        target_length = 13
    elif segment=='np':
        target_length = 8
    current_length = len(df)
    if current_length < target_length:
        filler_positions = range(current_length, target_length)
        filler_data = pd.DataFrame(0, index=filler_positions, columns=df.columns)
        df = pd.concat([df, filler_data])
    
    # reindex so there's no space between residues
    pos_mapping = {pos: i for i, pos in enumerate(resis)}
    df = df.rename(index=pos_mapping)
    

    
    logomaker.Logo(df, width=.8, vsep=0.001, color_scheme='hydrophobicity', 
                   font_name = 'Arial Rounded MT Bold', show_spines=False, ax=ax)
    
    xticks = [i for i in range(len(resis))]
    
    ax.set_xticks(xticks)
    ax.set_xticklabels(resis, rotation=45, ha='right', fontsize=10)

    
    # Set the x-axis limit to only span real residues and center it
    left_offset = (target_length - len(resis)) / 2
    ax.set_xlim(-left_offset, len(resis) - 1 + left_offset)
    
    ax.spines['bottom'].set_color('white')
#     ax.spines['bottom'].set_bounds(0, len(resis) - 1)

    
    # make y-axis invisible
    ax.yaxis.set_visible(False) 
    ax.spines['left'].set_color('white')
    
    plt.tight_layout()



#     ax.set_ylabel('Proportion of egg\nmutations at residue')
#     ax.set_xlabel('Residue')

    sns.despine()
                        
    
```

```python
def plot_fig3(threshold=0.03, filename=None):
    """
    Plot HA positions that commonly get mutations during egg-passaging
    Show where these are on the structure
    """

    fig = plt.figure(figsize=(10, 15))
    # include blank row to give more space underneath plots for axis labels 
    # (without having to increase space between strucutre and logo plot)
    # height order is structure, logo, blank space, blank space (logo gets the first half of the blank space too) 
    # (lollipop=structure+logo)
    gs = GridSpec(16, 2, width_ratios=[3,3], height_ratios=[5,0.6,0.3,0.2, 5,0.6,0.3,0.2, 
                                                            5,0.6,0.3,0.2, 5,0.6,0.3,0.2])

    
    # row 1 is H3N2
    # lollipop plot
    ax1 = fig.add_subplot(gs[0:2, 0])
    plot_egg_adaptive_muts('h3n2', 'ha', threshold, ax1)
    ax1.text(-0.2, 0.95, 'A', fontsize=20, transform=ax1.transAxes)
    
    # then structure
    ax2 = fig.add_subplot(gs[0, 1])
    img_h3n2 = mpimg.imread('Figure3/h3n2_HA_structure2.png')
    ax2.imshow(img_h3n2)
    ax2.axis('off')
#     ax2.text(-0.3, 0.9, 'B', fontsize=20, transform=ax2.transAxes)
    
    # with logo plot underneath
    ax2_1 = fig.add_subplot(gs[1:3, 1])
    make_logo_plot('h3n2', 'ha', 0.03, ax2_1)
#     ax2_1.text(-0.05, 1.15, 'C', fontsize=20, transform=ax2_1.transAxes)
    
    
    # row 2 is H1N1pdm
    # lollipop plot
    ax3 = fig.add_subplot(gs[4:6, 0])
    plot_egg_adaptive_muts('h1n1pdm', 'ha', threshold, ax3)
    ax3.text(-0.2, 0.95, 'B', fontsize=20, transform=ax3.transAxes)
    
    # structure
    ax4 = fig.add_subplot(gs[4, 1])
    img_h1n1pdm = mpimg.imread('Figure3/h1n1pdm_HA_structure2.png')
    ax4.imshow(img_h1n1pdm)
    ax4.axis('off')
#     ax4.text(-0.2, 0.9, 'E', fontsize=20, transform=ax4.transAxes)
    
    # with logo plot underneath
    ax4_1 = fig.add_subplot(gs[5:7, 1])
    make_logo_plot('h1n1pdm', 'ha', 0.03, ax4_1)
#     ax4_1.text(-0.05, 1.15, 'F', fontsize=20, transform=ax4_1.transAxes)

    
    # row 3 is Vic HA 
    # lollipop plot
    ax5 = fig.add_subplot(gs[8:10, 0])
    plot_egg_adaptive_muts('vic', 'ha', threshold, ax5)
    ax5.text(-0.2, 0.95, 'C', fontsize=20, transform=ax5.transAxes)
    
    # structure
    ax6 = fig.add_subplot(gs[8, 1])
    img_vic = mpimg.imread('Figure3/vic_HA_structure2.png')
    ax6.imshow(img_vic)
    ax6.axis('off')
#     ax6.text(-0.2, 0.9, 'H', fontsize=20, transform=ax6.transAxes)
    
    # with logo plot underneath
    ax6_1 = fig.add_subplot(gs[9:11, 1])
    make_logo_plot('vic', 'ha', 0.03, ax6_1)
#     ax6_1.text(0.3, 1.15, 'F', fontsize=20, transform=ax6_1.transAxes)

    
    # row 8 is Yam HA
    # lollipop plot
    ax7 = fig.add_subplot(gs[12:14, 0])
    plot_egg_adaptive_muts('yam', 'ha', threshold, ax7)
    ax7.text(-0.2, 0.95, 'D', fontsize=20, transform=ax7.transAxes)
    
    # structure
    ax8 = fig.add_subplot(gs[12, 1])
    img_yam = mpimg.imread('Figure3/yam_HA_structure2.png')
    ax8.imshow(img_yam)
    ax8.axis('off')
#     ax8.text(-0.2, 0.9, 'K', fontsize=20, transform=ax8.transAxes)
    
    # with logo plot underneath
    ax8_1 = fig.add_subplot(gs[13:15, 1])
    make_logo_plot('yam', 'ha', 0.03, ax8_1)
#     ax8_1.text(0.3, 1.15, 'L', fontsize=20, transform=ax8_1.transAxes)
    
#     plt.subplots_adjust(wspace=0.2, hspace=0.4)
    plt.subplots_adjust(wspace=0.2, hspace=0.3, top=0.95, bottom=0.05)
    
    
    if filename:
        fig.savefig(filename, dpi=300, bbox_inches='tight', transparent=False, facecolor='white')
    
    
```

```python
def get_num_sites_over_threshold(virus, threshold):
    """
    For all segments, find sites that are mutated in greater than the threshold proportion of egg-passaged strains
    """
    
    segments = ['pb1', 'pb2', 'pa', 'ha', 'np', 'na', 'mp', 'ns']
    
    count_over_threshold = {s:0 for s in segments}
    
    for s in segments:
        over_threshold, mut_pos_proportions, mut_pos_counts, total_strains = get_muts_over_threshold(virus, s, threshold)
        seg_count = 0
        for g, ps in over_threshold.items():
            seg_count+=len(ps)
        count_over_threshold[s] = seg_count
        
    return count_over_threshold
```

```python
def plot_fig3_old(threshold=0.03, filename=None):
    """
    Plot HA positions that commonly get mutations during egg-passaging
    Show where these are on the structure
    """

    fig = plt.figure(figsize=(10, 14))
    gs = GridSpec(4, 2, width_ratios=[3,2])

    
    # row 1 is H3N2
    ax1 = fig.add_subplot(gs[0, 0])
    plot_egg_adaptive_muts('h3n2', 'ha', threshold, ax1)
    ax1.text(-0.15, 0.95, 'A', fontsize=20, transform=ax1.transAxes)
    
    ax2 = fig.add_subplot(gs[0, 1])
    img_h3n2 = mpimg.imread('Figure3/h3n2_HA_structure.png')
    ax2.imshow(img_h3n2)
    ax2.axis('off')
    ax2.text(0, 0.95, 'B', fontsize=20, transform=ax2.transAxes)
    
    
    # row 2 is H1N1pdm
    ax3 = fig.add_subplot(gs[1, 0])
    plot_egg_adaptive_muts('h1n1pdm', 'ha', threshold, ax3)
    ax3.text(-0.15, 0.95, 'C', fontsize=20, transform=ax3.transAxes)
    
    ax4 = fig.add_subplot(gs[1, 1])
    img_h1n1pdm = mpimg.imread('Figure3/h1n1pdm_HA_structure.png')
    ax4.imshow(img_h1n1pdm)
    ax4.axis('off')
    ax4.text(0, 0.95, 'D', fontsize=20, transform=ax4.transAxes)
    
    
    # row 3 is Vic HA 
    ax5 = fig.add_subplot(gs[2, 0])
    plot_egg_adaptive_muts('vic', 'ha', threshold, ax5)
    ax5.text(-0.15, 0.95, 'E', fontsize=20, transform=ax5.transAxes)
    
    ax6 = fig.add_subplot(gs[2, 1])
    img_vic = mpimg.imread('Figure3/vic_HA_structure.png')
    ax6.imshow(img_vic)
    ax6.axis('off')
    ax6.text(0, 0.95, 'F', fontsize=20, transform=ax6.transAxes)
    
    # row 8 is Yam HA
    ax7 = fig.add_subplot(gs[3, 0])
    plot_egg_adaptive_muts('yam', 'ha', threshold, ax7)
    ax7.text(-0.15, 0.95, 'F', fontsize=20, transform=ax7.transAxes)
    
    ax8 = fig.add_subplot(gs[3, 1])
    img_yam = mpimg.imread('Figure3/yam_HA_structure.png')
    ax8.imshow(img_yam)
    ax8.axis('off')
    ax8.text(0, 0.95, 'G', fontsize=20, transform=ax8.transAxes)
    
#     plt.subplots_adjust(wspace=0.2, hspace=0.4)
    plt.subplots_adjust(wspace=0.2, hspace=0.3, top=0.95, bottom=0.05)
    
    if filename:
        fig.savefig(filename, dpi=300, bbox_inches='tight', transparent=False, facecolor='white')
    
    
```
